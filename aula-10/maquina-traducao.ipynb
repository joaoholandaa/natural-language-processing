```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model = AutoModelForSeq2SeqLM.from_pretrained('Helsinki-NLP/opus-mt-es-en')
tokenizer = AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-es-en')
```

```python
batch_input_str = (('Mary gasta $ 20 en pizza'),
		               ('A ella le gusta comerlo'),
		               ('La pizza estuvo genial'))
```

```python
encoded = tokenizer(batch_input_str, return_tensors='pt', padding=True)
translated = model.generate(**encoded)
tokenizer.batch_decode(translated, skip_special_tokens=True)
```

**Máquina de Tradução Multilíngue**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model = AutoModelForSeq2SeqLM.from_pretrained('Helsinki-NLP/opus-mt-en-ROMANCE')
tokenizer = AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-ROMANCE')
```

```python
tokenizer.supported_language_codes
```

```python
batch_input_str = (('>>pt_br<<Welcome to our class!'),
		               ('>>pt_br<<We are going to learn about Natural Language Processing.'),
		               ('>>pt_br<<We will also briefly talk about Machine Translation.'))
```

```python
encoded = tokenizer(batch_input_str, return_tensors='pt', padding=True)
translated = model.generate(**encoded)
tokenizer.batch_decode(translated, skip_special_tokens=True)
```

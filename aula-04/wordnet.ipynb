**WordNet no NLTK**

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import wordnet

wordnet.synsets('casa', lang='por')
```

```python
synsets_vela = wordnet.synsets('vela', lang='por')

vela1 = wordnet.synset('sailing.n.02').lemma_names('por')
vela2 = wordnet.synset('candle.n.01').lemma_names('por')
vela3 = wordnet.synset('spark_plug.n.01').lemma_names('por')

print(synsets_vela, '\n', vela1, '\n', vela2, '\n', vela3)
```

```python
hiper_vela1 = wordnet.synset('sailing.n.02').hypernyms()
hiper_vela2 = wordnet.synset('candle.n.01').hypernyms()
hiper_vela3 = wordnet.synset('spark_plug.n.01').hypernyms()

print(hiper_vela1, '\n', hiper_vela2, '\n', hiper_vela3)
```

```python
hipo_vela1 = wordnet.synset('sailing.n.02').hyponyms()
hipo_vela2 = wordnet.synset('candle.n.01').hyponyms()
hipo_vela3 = wordnet.synset('spark_plug.n.01').hyponyms()

print(hipo_vela1, '\n', hipo_vela2, '\n', hipo_vela3)
```
